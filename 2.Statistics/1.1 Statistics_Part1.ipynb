### 3. Distribution curve[Mean , Median ,mode ] --> Just understanding a data distribution 

```python
col = df["total_bill"]

mean_value = np.mean(col)
median_val = np.median(col)
mode_val = col.mode()[0]

print("Mean of total_bill:$", format(mean_value, ".2f"))
print("Median of total_bill:$", format(median_val, ".2f"))
print("Mode of total_bill:$", format(mode_val, ".2f"))

# visualizing the histogram of total_bill
plt.figure(figsize=(10, 6))
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean : {mean_value:.2f}')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f'Median : {median_val:.2f}')
plt.axvline(mode_val, color='blue', linestyle='dashed', linewidth=2, label=f'Mode : {mode_val:.2f}')
sns.histplot(col, kde=True, color='blue')
plt.title("Distribution of Total Bill Amounts")
plt.legend()
plt.show()
```

### 3. Measuring of Spread

Spread tell us how Scattered the data is around the center.

```python
#range
data_range = col.max() - col.min()

# variance 
variance = np.var(col)

# standard deviation
std_dev = np.sqrt(variance)

# IQR - Interquartile Range
Q1 = np.percentile(col, 25)
Q3 = np.percentile(col, 75)
IQR = Q3 - Q1

print("Range of total_bill:$", format(data_range, ".2f"))
print("Variance of total_bill:$", format(variance, ".2f"))
print("Standard Deviation of total_bill:$", format(std_dev, ".2f"))
print("Interquartile Range of total_bill:$", format(IQR, ".2f"))
```
