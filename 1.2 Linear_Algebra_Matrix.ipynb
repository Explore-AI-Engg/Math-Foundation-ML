### important operations on Matrix 

Matrix Additions and Subtraction can be added only if the have the same shape, that is both matrix should have the same no. of row and coloumn

Multiplication -- the no of coloumn in the first matrix should be equals to number of row in the second matrix 

importing Numpy Library

```python
import numpy as np
```

# Matrix Addition 

two Matrices can be added only if they have the row and coloumn (same shape)

```python
# creating the two Matrix

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("Matrix A:", A)
print("Matrix B:", B)
```

```python
A.shape
```

```python
B.shape
```

# Adding two Matrices 

```python
sum = A + B
print("Sum of A and B:", sum)
```

```python
# another way to add two matrix

matrix_1 = np.random.randint(10, size=(3, 3))
matrix_2 = np.random.randint(30, size=(3, 3))
print("Matrix 1:\n", matrix_1)
print("Matrix 2:\n", matrix_2)

matrix_sum = np.add(matrix_1, matrix_2)
print("Sum of Matrix 1 and Matrix 2:\n", matrix_sum)
```

# Subtraction of two Matrices 

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("Matrix A:", A)
print("Matrix B:", B)
```

```python
difference = A - B
print("Difference of A and B:", difference)
```

```python
matrix_1 = np.random.randint(10, size=(3, 3))
matrix_2 = np.random.randint(30, size=(3, 3))
print("Matrix 1:\n", matrix_1)
print("Matrix 2:\n", matrix_2)
```

```python
subtarction = np.subtract(matrix_1, matrix_2)
print("Subtraction of Matrix 1 and Matrix 2:\n", subtarction)
```

### Miltpliying 2 matrix 

```python
matrix3 = np.random.randint(10, size=(3, 3))
matrix4 = np.random.randint(30, size=(3, 5))
try:
    product = np.dot(matrix3, matrix4)
    print("Product of Matrix 3 and Matrix 4:\n", product)
except ValueError as e:
    print("Error:", e)
```

```python
matrix3 = np.random.randint(10, size=(3, 3))
matrix4 = np.random.randint(30, size=(4, 5))
try:
    product = np.dot(matrix3, matrix4)
    print("Product of Matrix 3 and Matrix 4:\n", product)
except ValueError as e:
    print("Error:", e)
```
